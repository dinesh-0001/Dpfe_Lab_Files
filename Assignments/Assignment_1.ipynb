<h3>Filling the None Values With the Mean, Mode and Median</h3>

```python
mean_window=df4['Window'].mean()
```

```python
mean_window
```

```python
import numpy as np #Importing Numpy
```

```python
df4['Window']=df4['Window'].fillna(mean_window)
```

```python
df4
```

```python
df4['Window']=df4['Window'].apply(np.ceil) #Used to round off the 1.75 Value to 2 because value of window can't be in like this
```

```python
df4
```

```python
median_Door=df4['Door'].median()
```

```python
median_Door
```

```python
df4['Door']=df4['Door'].fillna(median_Door)
```

```python
df4
```

```python
df4['Fans']=df4['Fans'].cat.codes
```

```python
df4
```

```python
mean_Switch_Board=df4['Switch_Board'].mean()
```

```python
mean_Switch_Board
```

```python
df4['Switch_Board']=df4['Switch_Board'].fillna(mean_Switch_Board)
```

```python
df4
```

```python
df4['Switch_Board']=df4['Switch_Board'].apply(np.floor) #Make the 2.25 to floor Value 2
```
